```python
## from 1 hour to 2 hours
import matplotlib.pyplot as plt;
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6];
z=[3,2,4,5,1,6,2];
color=['r','y','g','b','r','g','r']
size=[10,20,30,40,50,60,70]
plt.scatter(x,y,c=color,s=size,
            alpha=0.5,
            marker="+",
            edgecolors='g',
            linewidths=4,
            cmap='viridis'
            );
plt.scatter(x,z,c=color,s=size,
            alpha=0.5,
            marker="+",
            edgecolors='g',
            linewidths=4,
            cmap='viridis'
            );
t=plt.colorbar()
t.set_label('ColorBar',fontsize=15)
plt.title("scatter plot")
plt.xlabel("x-label",fontsize=15)
plt.ylabel("y-label",fontsize=15)

plt.show();
```

```python
import matplotlib.pyplot as plt;
x=['Python','c','C++','Javascript'];
y=[85,70,60,80];
z=[20,30,40,50]
c=['y','b','m','g'];
plt.bar(x,y,width=0.5,color=c,
        align='center',
        edgecolor='r',
        linewidth=5,
        linestyle=":",
        alpha=0.5,label="rifat");
plt.bar(x,z,width=0.5,color=c,
        align='center',
        edgecolor='y',
        linewidth=5,
        linestyle=":",
        alpha=0.5,label="rifat");
plt.legend();

plt.xlabel('programming Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)
plt.title("Language Popularity")
plt.show();
```

```python
import matplotlib.pyplot as plt;
import numpy as np;

x=['Python','c','C++','Javascript'];
y=[85,70,60,80];
z=[20,30,40,50]
c=['y','b','m','g'];
p=np.arange(len(x))
w=0.2

p1=[j+w for j in p ]

plt.bar(p,y,width=w,color=c,
        align='center',
        edgecolor='r',
        linewidth=5,
        alpha=0.5,label="rifat");
plt.bar(p1,z,width=w,color=c,
        align='center',
        edgecolor='y',
        linewidth=5,
        alpha=0.5,label="rifat");

plt.yticks();
plt.legend();

plt.xlabel('programming Language',fontsize=10)
plt.ylabel('Popularity',fontsize=15)
plt.title("Language Popularity")
plt.xticks(p+w/2,x,rotation=20);
plt.show();
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'C', 'C++', 'Javascript']
y = [85, 70, 60, 80]  # First dataset
z = [20, 30, 40, 50]  # Second dataset
colors = ['y', 'b', 'm', 'g']
positions = np.arange(len(x))
width = 0.2

plt.barh(positions, y, height=width, color=colors,
         align='center',
         edgecolor='r',
         linewidth=2,
         alpha=0.7, label="Popularity")

plt.barh(positions + width, z, height=width, color=colors,
         align='center',
         edgecolor='y',
         linewidth=2,
         alpha=0.7, label="Another Metric")

plt.yticks(positions + width/2, x)
plt.legend()

plt.xlabel('Value', fontsize=12)
plt.ylabel('Programming Language', fontsize=12)
plt.title("Language Popularity Comparison")
plt.tight_layout()
plt.show()
```
