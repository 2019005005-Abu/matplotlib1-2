```python
'''Matplotlib Pie Plot'''
import matplotlib.pyplot as plt;
x=[10,20,30,40]
y=["c","c+","Js","Python"]
c=['r','g','b','y']
ex=[0.0,0.0,0.0,0.2]
plt.pie(x,labels=y,
        explode=ex,
        colors=c,
        autopct="%0.1f%%",
        shadow=True,
        radius=1.5,
        labeldistance=1.3,
        startangle=90,
        textprops={"fontsize":15},
        counterclock=False,
        wedgeprops={
            "linewidth":5,
            "edgecolor":"m"
            },
        center=(1,3),
        rotatelabels=False
        );
plt.title("wscube tech")
plt.legend(loc=3);
plt.show();

```

```python
x=[10,20,30,40]
y=["c","c+","Js","Python"]
c=['r','g','b','y']

plt.pie(x,labels=y,radius=1.5);
cr=plt.Circle(xy=(0,0),
              radius=1,
              facecolor="w",
              );
plt.gca().add_artist(cr);
plt.show();
```

```python
'''Matplotlib Stem Plot'''
import matplotlib.pyplot as plt;
x=[1,2,3,4,5,6];
y=[2,2,5,7,3,8];
plt.stem(x,y);
plt.show();
```

```python
x=[1,2,3,4,5,6];
y=[2,2,5,7,3,8];
plt.stem(x,y,
         linefmt=":",
         markerfmt='r+',
         bottom=1,
         basefmt="g",
         label="python",
         );
plt.legend();
plt.title("Without Horizontal")
plt.show();
x=[1,2,3,4,5,6];
y=[2,2,5,7,3,8];
plt.stem(x,y,
         linefmt=":",
         markerfmt='r+',
         bottom=1,
         basefmt="g",
         label="python",
         orientation="horizontal"
         );
plt.legend();
plt.show();
```

```python
##Box vs Whisker Plot##
x=[10,20,30,40,50,60,70,80,200]
y=['c','c++','R','python','Js']

plt.boxplot(x,
            # notch=True,
            vert=True,
            widths=0.5,
            labels=['python'],
            patch_artist=True,
            showmeans=True,
            whis=2,
            sym='g+',
            boxprops=dict(color="r"),
            capprops=dict(color="g"),
            whiskerprops=dict(color='r'),
            
            )
plt.show();
```

```python
##Box vs Whisker Plot##
x=[10,20,30,40,50,60,70,80,200]
x1=[10,20,40,50,30,60,90,100,110]
y=[x,x1];
plt.boxplot(y,
            # notch=True,
            vert=True,
            labels=['python','c++'],
            patch_artist=True,
            showmeans=True,
            whis=2,
            sym='g+',
            boxprops=dict(color="r"),
            capprops=dict(color="g"),
            whiskerprops=dict(color='r'),
            )
plt.show();


```
