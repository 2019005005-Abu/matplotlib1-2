```python
'''Bar Plot'''
import matplotlib.pyplot as plt;
import numpy as np;
x=[i for i in range(1,5)];
y=[j for j in range(5,9)]
x1=['Python','c','c++','java'];
y1=[95,50,60,90];
z1=[45,40,30,51]
w=.4;
p=np.arange(len(x1));
p1=[j+w for j in p]
plt.xlabel('language',fontsize="10");
plt.ylabel('point',fontsize="15");
plt.title("WSCUBE")

c='b'
mc=['r','g','b','m']
bar__list=[0,1,2,3];
textAlign='center'
edgeColorIs='r'
LineWidth=1
LIneStyle=':'
al=0.5
LB='popularity1'
LB1='popularity2'
plt.bar(x1,y1,width=w,align=textAlign,label=LB);
plt.show();
plt.bar(p,y1,color=mc,width=w,align=textAlign,edgecolor=edgeColorIs,linewidth=LineWidth,linestyle=LIneStyle,alpha=al,label=LB);
plt.bar(p1,z1,color='y',width=w,align=textAlign,edgecolor=edgeColorIs,linewidth=LineWidth,linestyle=LIneStyle,alpha=al,label=LB1);
plt.xticks(p+w/2,x1,rotation=20);
# plt.yticks();
plt.legend();
plt.show()
```
